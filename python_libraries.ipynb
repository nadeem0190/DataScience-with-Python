```python
name = ['a','b','c','d','e']
salary = np.random.randint(15000,50000,5)
```

```python
salary
```

```python
plt.bar(name,salary, color = 'green')
plt.xlabel('Name')
plt.ylabel('Salary')
plt.title('Name x Salary')
plt.show()
```

```python
color1 = ['r','g','b','#caeb91','#5a8491']
plt.bar(name,salary, color = color1)
plt.xlabel('Name')
plt.ylabel('Salary')
plt.title('Name x Salary')
plt.show()
```

```python
color1 = ['r','g','b','#caeb91','#5a8491']
plt.barh(name,salary, color = color1)
plt.xlabel('Name')
plt.ylabel('Salary')
plt.title('Name x Salary')
plt.show()
```
